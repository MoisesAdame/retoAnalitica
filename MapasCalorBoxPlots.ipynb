## Diagrama de Cajas y Bigotes e Histogramas
### user_followers

```python
fig, ax = plt.subplots(1, 3)
ax[0].hist(df['user_followers'], range=[0.0 * 10**7, 0.2 * 10**7])
ax[0].set_title("Histograma \nRango = (0, 0.2) 10^7")

ax[1].hist(df['user_followers'], range=[0.2 * 10**7, 1.4 * 10**7])
ax[1].set_title("Histograma \nRango = (0.2, 1.4) 10^7")

ax[2].boxplot(df['user_followers'])
ax[2].set_title("Diagrama de \nCajas y Bigotes")

fig.suptitle('user_followers')
fig.tight_layout(pad=0.5)
plt.subplots_adjust(wspace=0.5)

plt.show()
```

### user_friends

```python
fig, ax = plt.subplots(1, 3)
ax[0].hist(df['user_friends'], range=[0, 10**3])
ax[0].set_title("Histograma \nRango = (0, 1) 10^3")

ax[1].hist(df['user_friends'], range=[10**3, 10**4])
ax[1].set_title("Histograma \nRango = (1, 10) 10^3")

ax[2].boxplot(df['user_friends'])
ax[2].set_title("Diagrama de \nCajas y Bigotes")

fig.suptitle('user_friends')
fig.tight_layout(pad=0.5)
plt.subplots_adjust(wspace=0.5)

plt.show()
```

### user_favourites

```python
fig, ax = plt.subplots(1, 2)
ax[0].hist(df['user_favourites'], range = [0, 0.5 * 10 ** 5])
ax[0].set_title("Histograma \nRango = (0, 0.5) 10^5")

ax[1].boxplot(df['user_favourites'])
ax[1].set_title("Diagrama de \nCajas y Bigotes")

fig.suptitle('user_friends')

plt.show()
```
